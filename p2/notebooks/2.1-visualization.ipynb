```python
from sqlite3 import connect
import numpy as np
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
conditions_path = '../data/raw/synthea/scenario1/conditions.csv'
encounters_path = '../data/raw/synthea/scenario1/encounters.csv'
patients_path = '../data/raw/synthea/scenario1/patients.csv'
procedures_path = '../data/raw/synthea/scenario1/procedures.csv'
```

```python
conn = connect(':memory:')
```

```python
conds_df = pd.read_csv(conditions_path)
conds_df.to_sql('conditions', conn, if_exists='replace')
conds_df.head()
```

```python
encouters_df = pd.read_csv(encounters_path)
encouters_df.to_sql('encounters', conn, if_exists='replace')
encouters_df.head()
```

```python
procedures_df = pd.read_csv(procedures_path)
procedures_df['DEPRESSION_SCREENING'] = procedures_df['DESCRIPTION'].apply(lambda x: 'depression' in x.lower())
procedures_df.to_sql('procedures', conn, if_exists='replace')
procedures_df.head()
```

```python
patients_df = pd.read_csv(patients_path)
patients_df.to_sql('patients', conn, if_exists='replace')
patients_df.head()
```

```python
patient_cond_df = pd.read_sql(
    '''
    SELECT p.Id, p.birthdate, p.deathdate, c.description, c.code
    FROM patients p
    JOIN conditions c on p.Id = c.patient;
    ''',
    conn
)
patient_cond_df.to_sql('patientCondition', conn, if_exists='replace')
patient_cond_df.head()
```

```python
def join_with_comma(
    x
):
    return ','.join(str(y) for y in x)

grouped_proced_df = procedures_df.sort_values(by='START').groupby('PATIENT').aggregate(
    {
        'CODE': join_with_comma,
        'START': lambda x: np.diff(x.astype(np.datetime64)).mean(),
        'DESCRIPTION': join_with_comma,
        'REASONCODE': join_with_comma,
        'REASONDESCRIPTION': join_with_comma,
        'DEPRESSION_SCREENING': sum,
    }
)
```

```python
grouped_proced_df.head()
```

```python
def join_with_comma(
    x
):
    return ','.join(str(y) for y in x)

grouped_depressive_df = procedures_df[procedures_df['DEPRESSION_SCREENING']].sort_values(by='START').groupby('PATIENT').aggregate(
    {
        'START': lambda x: np.diff(x.astype(np.datetime64)).mean() / np.timedelta64(1, 'D'),
        'DEPRESSION_SCREENING': 'sum',
    }
)
grouped_depressive_df.rename(columns={'START': 'INTERVAL'}, inplace=True)
grouped_depressive_df.to_sql('depressive', conn, if_exists='replace')
grouped_depressive_df.head()
```

```python
patient_depressive = pd.read_sql(
    '''
    SELECT p.Id, p.birthdate, p.deathdate,d.interval, d.depression_screening
    FROM patients p
    JOIN depressive d on p.Id = d.patient;
    ''',
    conn
)
patient_depressive.head()
```
