```python
def join_with_comma(
    x
):
    return ','.join(str(y) for y in x)
```

```python
grouped_proced_df = procedures_df.sort_values(by='START').groupby('PATIENT').aggregate(
    {
        'START': lambda x: np.diff(x.astype(np.datetime64)).mean() / np.timedelta64(1, 'D'),
        'DEPRESSION_SCREENING': 'sum',
    }
)
grouped_proced_df.rename(columns={'START': 'INTERVAL'}, inplace=True)
grouped_proced_df.to_sql('procedures', conn, if_exists='replace')
grouped_proced_df.head()
```

```python
grouped_encount_df = encounters_df.sort_values(by='STOP').groupby('PATIENT').aggregate(
    {
        'START': 'last',
        'STOP': 'last',
        'CODE': 'last',
        'REASONCODE': 'last',
    }
)
grouped_encount_df['DURATION'] = (grouped_encount_df['STOP'].astype(np.datetime64) - grouped_encount_df['START'].astype(np.datetime64)) / np.timedelta64(1, 'D')
grouped_encount_df.to_sql('encounters', conn, if_exists='replace')
grouped_encount_df.head()
```

```python
patient_depressive = pd.read_sql(
    '''
    SELECT p.Id, p.birthdate, p.deathdate, d.interval, d.depression_screening, e.code, e.reasoncode, e.stop, e.duration
    FROM patients p
    JOIN procedures d on p.Id = d.patient
    JOIN encounters e on p.Id = e.patient
    ''',
    conn
)
patient_depressive.rename(columns={
    'CODE': 'ENCOUNTER_CODE',
    'REASONCODE': 'ENCOUNTER_REASONCODE',
    'STOP': 'ENCOUNTER_STOP',
    'DURATION': 'ENCOUNTER_DURATION',
}, inplace=True)
patient_depressive.head()
```
