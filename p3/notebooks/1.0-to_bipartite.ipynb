```python
import pandas as pd
import numpy as np
```

```python
graph = pd.read_csv('../data/raw/comorbidity_odds_matrix.csv')
```

```python
graph
```

```python
## transform the graph into a bipartite graph
bipartite = pd.DataFrame(
    data=np.zeros((graph.shape[0]*(graph.shape[0]-1), 3)),
    columns=['source', 'target', 'weight']
)

b_idx = 0
for i in range(graph.shape[0]):
    for j in range(1,graph.shape[1]):
        if graph.iloc[i, j] != 0:
            bipartite.loc[b_idx] = pd.Series({
                'source': graph.loc[i, 'Unnamed: 0'],
                'target': graph.columns[j],
                'weight': graph.iloc[i, j]
            })
            b_idx += 1
```
